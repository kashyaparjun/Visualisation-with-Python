# Plots and Subplots

This module deals with customising the Matplotlib library to suit the type of visuals that you want to produce.

```python
# Let's import Pyplot
%matplotlib inline
import matplotlib.pyplot as plt
```

In the previous tutorial we saw how we can produce a straight line graph using the plot function of Pyplot. Now we will modify it to produce different visuals.

The plot() function takes a parameter fmt which is nothing but the format in which the graph has to be visualised. It is 'b-' by default, where 'b' is for the blue color, and '-' is for the line. You can see the different available formats [here](https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.pyplot.plot.html).

Now, we shall format the graph to use blue triangle-up marker, the format for it is 'b^'.

```python
x = [1,2,3,4]
y = [2,3,4,5]
```

```python
plt.plot(x,y,'b^')
plt.ylabel('This is y axis')
plt.xlabel('This is x axis')
plt.show()
```

Similarly, we can plot a graph with format of red square marker using the format 'rs'

```python
plt.plot(x,y,'rs')
plt.ylabel('This is y axis')
plt.xlabel('This is x axis')
plt.show()
```

We can also plot sqaured and cubic functions using the plot function.

We can pass the arguments in the format: x1, y1, fmt1, x2, y2, fmt2, x3, y3, fmt3, and so on...

Here, we will prepare y1 and y2 which is y^2 and y3 which is y^3, the x can remain the same throughout. 

1. For the plot x, y we shall use the format 'mh' for magenta color and hexagon marker
2. For the plot x, y^2 we shall use the format 'b+' for blue color and plus marker
3. For the plot x, y^3 we shall use the format 'gv' for green color and a triangle down marker

```python
x = [i for i in range(10)]
y = [i for i in range(10)]
y2 = [i**2 for i in y]
y3 = [i**3 for i in y]
```

```python
plt.plot(x, y, 'mh', x, y2, 'b+', x, y3, 'gv')
plt.show()
```

### Subplots

In the above graph we have 3 different plots which have a similar x axis due to which we are able to add all three of them into one single graph. In the case we would have separate the graphs but display them as part of the same image, we use subplots.

We initialise the plt.figure() function which helps us create subplots in a single figure. We can pass the figsize or figure size as parameter to the figure functions. For example, plt.figure(figsize=(5,10)), where 5 in the width and 10 is the height in inches.

The subplot function takes a three digit number. For example, plt.subplot(211), where 2 is 2 rows, 1 column, and 1 specifies the position of the current graph.

```python
plt.figure(figsize=(15,8))

# plotting the x and y plot
plt.subplot(131)
plt.plot(x,y,'mh')
plt.xlabel('x')
plt.ylabel('y')

# plotting the x and y^2 plot
plt.subplot(132)
plt.plot(x,y2,'b+')
plt.xlabel('x')
plt.ylabel('y^2')

# plotting the x and y^3 plot
plt.subplot(133)
plt.plot(x,y3,'gv')
plt.xlabel('x')
plt.ylabel('y^3')

plt.show()
```
